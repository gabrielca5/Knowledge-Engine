```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/naruto.csv", encoding='latin1', index_col=0)

df.tail()
```

```python

plt.figure(figsize=(15, 7))
sns.boxplot(data=df, x="Saga", y="Rate")

plt.xticks(rotation=270)
plt.tight_layout()
plt.show()
```

## Boxplot das notas por saga

Conseguimos tirar alguns insights do boxplot, vemos que, em geral, as notas se mantÃ©m acima de 6, o que Ã© um bom sinal, mas tambÃ©m vemos que existem algumas notas abaixo disso, o que pode indicar que existem alguns filmes que nÃ£o foram tÃ£o bem recebidos pelos espectadores. AlÃ©m disso, podemos observar que a maioria das notas estÃ£o concentradas entre 7 e 8, o que sugere que a maioria dos filmes da saga sÃ£o bem avaliados.

Boxplots mais longos, indicam uma maior variaÃ§Ã£o nas notas, ou seja, indica que o arco pode ter partes boas e outras partes ruins.
